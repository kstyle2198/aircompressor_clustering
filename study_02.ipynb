```python
import numpy as np
import math as m
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import silhouette_score
import warnings
warnings.filterwarnings(action='ignore') 
```

```python
train_data = pd.read_csv("./dataset/train_data.csv")
features = ["air_end_temp", "motor_current"]
X = train_data[features]
X.head(2)
```

```python
fig = px.scatter(X, x="air_end_temp", y="motor_current")
fig.show()
```

```python
mins = 8
nn = NearestNeighbors(n_neighbors=mins+1, radius=0.4)
nn.fit(X[['air_end_temp', 'motor_current']])
distances, neighbors = nn.kneighbors(X[['air_end_temp', 'motor_current']])
```

```python
distances
```

```python
neighbors
```

```python
distances = np.sort(distances[:, mins], axis = 0)
distances
```

```python
distance_df = pd.DataFrame({"distances": distances, 
                           "index": list(range(0, len(distances)))})
```

```python
fig = px.scatter(distance_df, x="index", y="distances")
fig.show()
```
